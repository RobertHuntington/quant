```python
import pandas
```

```python
btc_usdt_path = populator.get_data_path(data_dir, 'binance', 'BTC/USDT', tick_size, start)
btc_usdt_frame = pandas.read_csv(btc_usdt_path, index_col=0)
btc_usdt_frame.index = pandas.to_datetime(btc_usdt_frame.index, unit='ms')
```

```python
btc_usdt_frame
```

```python
import backtrader as bt
import backtrader.analyzers as analyzers

class SimpleRsiStrategy(bt.Strategy):
    def __init__(self):
        self.rsi = bt.indicators.RSI_SMA(self.data.close, period=21)

    def next(self):
        if not self.position:
            if self.rsi < 30:
                self.buy(size=10)
        else:
            if self.rsi > 70:
                self.sell(size=10)

cerebro = bt.Cerebro()
cerebro.addstrategy(SimpleRsiStrategy)
data = bt.feeds.PandasData(dataname=btc_usdt_frame)
cerebro.adddata(data)

cerebro.addanalyzer(analyzers.SharpeRatio, _name='sharpe')
cerebro.addanalyzer(analyzers.DrawDown, _name='drawdown')

starting_cash = 100000
cerebro.broker.setcash(starting_cash)
strat = cerebro.run()

# End diagnostics.
portfolio_value = cerebro.broker.getvalue()
pnl = portfolio_value - starting_cash

print('Final Portfolio Value: ${}'.format(portfolio_value))
print('P/L: ${}'.format(pnl))

print('Sharpe Ratio: {}'.format(strat[0].analyzers.sharpe.get_analysis()))
print('Drawdown Stats: {}'.format(strat[0].analyzers.drawdown.get_analysis()))
```
